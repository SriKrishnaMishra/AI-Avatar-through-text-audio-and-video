STEP 4: Initialize Chatbot System

---

```python

import torch

# Try to import TTS, fallback to gTTS if fails
try:
    from TTS.api import TTS
    use_xtts = True
    print("✅ Using XTTS (advanced voice cloning)")
except:
    from gtts import gTTS
    use_xtts = False
    print("⚠️ Using gTTS (basic text-to-speech)")

# Initialize TTS
if use_xtts:
    try:
        device = "cuda" if torch.cuda.is_available() else "cpu"
        tts_model = TTS(model_name="tts_models/multilingual/multi-dataset/xtts_v2",
                        gpu=(device=="cuda"))
        print(f"✅ XTTS loaded on {device}")
    except Exception as e:
        print(f"⚠️ XTTS failed: {e}")
        use_xtts = False
        print("⚠️ Falling back to gTTS")

# Knowledge Base - 20 Q&A pairs
qa_database = {
    "what is this platform": "This is an LMS platform for online learning.",
    "how can i enroll": "You can enroll by clicking the enroll button on any course.",
    "is this free": "Some courses are free, others require payment.",
    "reset password": "Use the forgot password option on the login page.",
    "download certificate": "Yes, certificates are available after course completion.",
    "payment methods": "We accept UPI, credit cards, debit cards, and net banking.",
    "mobile access": "Yes, our platform works on all mobile devices.",
    "course access": "Most courses offer lifetime access after enrollment.",
    "beginner friendly": "Yes, we have many courses designed for beginners.",
    "track progress": "Yes, you can track your progress in your dashboard.",
    "support": "We provide 24 by 7 customer support via chat and email.",
    "cancel subscription": "Yes, you can cancel your subscription anytime.",
    "live classes": "Some courses include live classes with instructors.",
    "recommendations": "Yes, we recommend courses based on your interests.",
    "data security": "Your data is encrypted and completely secure.",
    "instructor payment": "Instructors earn revenue per student enrollment.",
    "quizzes": "Yes, most courses include quizzes and assessments.",
    "contact support": "Click the help button or email support at example dot com.",
    "deadlines": "Most courses are self paced with no strict deadlines.",
    "languages": "We offer courses in English, Hindi, and Spanish."
}

def get_answer(question):
    """Match question to answer using keywords"""
    question = question.lower()

    for key, answer in qa_database.items():
        if key in question:
            return answer

    return "I'm still learning. Please try rephrasing your question."

print(f"✅ Knowledge base: {len(qa_database)} Q&A pairs loaded")
print("\n✅ CHATBOT INITIALIZED\n")
```

```python
def generate_speech(text, output_file="response.wav"):
    """Generate speech from text"""
    print(f"🗣️  Generating speech...")

    try:
        if use_xtts and voice_file and os.path.exists(voice_file):
            # XTTS with voice cloning
            tts_model.tts_to_file(
                text=text,
                speaker_wav=voice_file,
                language="en",
                file_path=output_file
            )
            print(f"✅ Speech generated with voice cloning")
        elif use_xtts:
            # XTTS with default voice
            tts_model.tts_to_file(
                text=text,
                language="en",
                file_path=output_file
            )
            print(f"✅ Speech generated with default voice")
        else:
            # gTTS fallback
            tts_obj = gTTS(text=text, lang='en', slow=False)
            tts_obj.save(output_file)
            print(f"✅ Speech generated with gTTS")

        if os.path.exists(output_file):
            size = os.path.getsize(output_file)
            print(f"   Audio file: {output_file} ({size} bytes)")
            return output_file
        else:
            raise Exception("Audio file not created")

    except Exception as e:
        print(f"❌ TTS Error: {e}")
        # Last resort: use gTTS
        try:
            print("🔄 Trying gTTS as fallback...")
            tts_obj = gTTS(text=text, lang='en', slow=False)
            tts_obj.save(output_file)
            print(f"✅ Speech generated with gTTS fallback")
            return output_file
        except Exception as e2:
            print(f"❌ All TTS methods failed: {e2}")
            return None
```

```python
def generate_video(audio_file, output_file="output.mp4"):
    """Generate video: static image + audio (100% reliable)"""
    print("🎬 Generating talking avatar video...")

    try:
        # Check files exist
        if not os.path.exists(avatar_file):
            raise FileNotFoundError(f"Avatar not found: {avatar_file}")
        if not os.path.exists(audio_file):
            raise FileNotFoundError(f"Audio not found: {audio_file}")

        # Get audio duration
        import subprocess
        duration_cmd = f"ffprobe -v error -show_entries format=duration -of default=noprint_wrappers=1:nokey=1 {audio_file}"
        duration = subprocess.check_output(duration_cmd, shell=True).decode().strip()
        print(f"   Audio duration: {float(duration):.1f} seconds")

        # Create video with ffmpeg (static image + audio)
        cmd = f"""ffmpeg -loop 1 -i {avatar_file} -i {audio_file} \
            -c:v libx264 -tune stillimage -c:a aac -b:a 192k \
            -pix_fmt yuv420p -shortest -y {output_file}"""

        print("   Creating video...")
        result = os.system(cmd + " > /dev/null 2>&1")

        if os.path.exists(output_file):
            size = os.path.getsize(output_file)
            print(f"✅ Video created: {output_file} ({size} bytes)")
            return output_file
        else:
            raise Exception("Video file not created")

    except Exception as e:
        print(f"❌ Video generation error: {e}")
        return None
```

```python
def chatbot(question):
    """Complete chatbot pipeline"""
    print("\n" + "="*70)
    print(f"💬 QUESTION: {question}")
    print("="*70 + "\n")

    # 1. Get answer from knowledge base
    answer = get_answer(question)
    print(f"🤖 ANSWER: {answer}\n")

    # 2. Generate speech
    audio_file = generate_speech(answer)
    if not audio_file:
        print("❌ Failed to generate audio")
        return None

    # 3. Generate video
    video_file = generate_video(audio_file, f"output_{len(os.listdir('.'))+1}.mp4")
    if not video_file:
        print("❌ Failed to generate video")
        return None

    print("\n" + "="*70)
    print(f"✅ SUCCESS! Video ready: {video_file}")
    print("="*70 + "\n")

    return video_file
```
